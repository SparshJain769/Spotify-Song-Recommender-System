```python
import os
import numpy as np
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from sentence_transformers import SentenceTransformer
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Initialize the Spotify client
client_id ='1c24933212744aaead28a35b58f2d1eb'
client_secret = '85f9b77d4e1445c3a761ec9871fce0eb'
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

def get_artist_details(artist_name):
    try:
        results = sp.search(q='artist:' + artist_name, type='artist')
        artists = results['artists']['items']
        if len(artists) > 0:
            artist = artists[0]
            albums = sp.artist_albums(artist['id'], album_type='album')
            top_tracks = sp.artist_top_tracks(artist['id'])
            return {
                'follower_count': artist['followers']['total'],
                'popularity': artist['popularity'],
                'genres': artist['genres'],
                'album_count': len(albums['items']),
                'top_track_count': len(top_tracks['tracks'])
            }
        else:
            return {'follower_count': 0, 'popularity': 0, 'genres': [], 'album_count': 0, 'top_track_count': 0}
    except Exception as e:
        print(f"Error fetching details for {artist_name}: {str(e)}")
        return {'follower_count': 0, 'popularity': 0, 'genres': [], 'album_count': 0, 'top_track_count': 0}

def get_album_details(album_id):
    try:
        album = sp.album(album_id)
        tracks = sp.album_tracks(album_id)
        return {
            'release_year': int(album['release_date'][:4]),
            'track_count': len(tracks['items']),
            'album_popularity': album.get('popularity', 0)
        }
    except Exception as e:
        print(f"Error fetching details for album ID {album_id}: {str(e)}")
        return {'release_year': 0, 'track_count': 0, 'album_popularity': 0}
```

```python
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer

def batch_encode(model, texts, batch_size=32):
    embeddings = []
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i + batch_size]
        batch_embeddings = model.encode(batch_texts)
        embeddings.append(batch_embeddings)
    return np.vstack(embeddings)

def create_enhanced_embeddings(df, batch_size=32):
    model = SentenceTransformer('distiluse-base-multilingual-cased-v2')  # Using a smaller model

    # Generate embeddings with batching
    song_name_embeddings = batch_encode(model, df['song_title'].tolist(), batch_size)
    artist_name_embeddings = batch_encode(model, df['artist_name'].tolist(), batch_size)
    # album_name_embeddings = batch_encode(model, df['album_id'].tolist(), batch_size)

    # Get additional artist and album details
    artist_details = df.apply(lambda row: get_artist_details(row['artist_name']), axis=1)
    album_details = df.apply(lambda row: get_album_details(row['album_id']), axis=1)

    # Extract and normalize numerical features
    artist_followers = np.log1p(artist_details.apply(lambda x: x['follower_count']))
    artist_popularity = artist_details.apply(lambda x: x['popularity'])
    artist_album_count = np.log1p(artist_details.apply(lambda x: x['album_count']))
    artist_top_track_count = artist_details.apply(lambda x: x['top_track_count'])
    album_track_count = album_details.apply(lambda x: x['track_count'])
    album_popularity = album_details.apply(lambda x: x['album_popularity'])

    # One-hot encode genres
    all_genres = set([genre for details in artist_details for genre in details['genres']])
    genre_matrix = np.zeros((len(df), len(all_genres)))
    for i, details in enumerate(artist_details):
        for genre in details['genres']:
            genre_matrix[i, list(all_genres).index(genre)] = 1

    return {
        'song_name_embeddings': song_name_embeddings,
        'artist_name_embeddings': artist_name_embeddings,
        'artist_followers': artist_followers,
        'artist_popularity': artist_popularity,
        'artist_album_count': artist_album_count,
        'artist_top_track_count': artist_top_track_count,
        'album_track_count': album_track_count,
        'album_popularity': album_popularity,
        'genre_matrix': genre_matrix
    }
```

COSINE SIMILARITIES

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute cosine similarity matrix
cosine_sim_matrix = cosine_similarity(normalized_features)

# Convert cosine similarity to cosine distance
cosine_dist_matrix = 1 - cosine_sim_matrix
```

```python
# Create a DataFrame for easier interpretation
song_names = df['Track Name'].values
cosine_sim_df = pd.DataFrame(cosine_sim_matrix, index=song_names, columns=song_names)
cosine_dist_df = pd.DataFrame(cosine_dist_matrix, index=song_names, columns=song_names)
```

```python
def get_cosine_metrics_by_index(index1, index2):
    if index1 >= len(song_names) or index2 >= len(song_names):
        raise ValueError(f"Indices {index1} and {index2} are out of range")

    song1 = song_names[index1]
    song2 = song_names[index2]

    sim = cosine_sim_df.iloc[index1, index2]
    dist = cosine_dist_df.iloc[index1, index2]
    return song1, song2, sim, dist

# Example usage
index1 = 26
index2 = 27
song1, song2, similarity, distance = get_cosine_metrics_by_index(index1, index2)
print(f"Cosine Similarity between '{song1}' and '{song2}': {similarity}")
print(f"Cosine Distance between '{song1}' and '{song2}': {distance}")
```

```python
def find_top_similar_pairs(similarity_matrix, song_names, top_n=5):
    # Get the upper triangle of the similarity matrix, excluding the diagonal
    upper_triangle_indices = np.triu_indices_from(similarity_matrix, k=1)

    # Get the similarity values from the upper triangle
    similarity_values = similarity_matrix[upper_triangle_indices]

    # Get the sorted indices of the top N similarity values
    sorted_indices = np.argsort(similarity_values)[-top_n:][::-1]

    # Get the corresponding song pairs and their similarity values
    top_pairs = [(song_names[upper_triangle_indices[0][i]],
                  song_names[upper_triangle_indices[1][i]],
                  similarity_values[i]) for i in sorted_indices]

    return top_pairs

# Get the top 5 similar song pairs
top_similar_pairs = find_top_similar_pairs(cosine_sim_df.values, song_names, top_n=5)

# Print the top similar song pairs
print("Top similar song pairs:")
for song1, song2, similarity in top_similar_pairs:
    print(f"'{song1}' and '{song2}' with similarity: {similarity}")
```
